## <font color="#004D7F">3.6. Correlaciones: _corr()_ </font>

Para saber como se relacion los atributos entre si (incluido el atributo clase) usamos una matriz de correlación de Pearson, para ello usamos la función `corr()` , esta es una matriz simétrica que tiene como diagonal principal a la unidad. 
Si el número en la matriz donde se intersecan esos 2 atributos es positivo quiere decir que se relacionan de manera directa, si es negativo lo hacen de forma inversa
Podemos decir que 2 atributos tienen una fuerte correlación si este número es >0.75 o <0.75. Aquí podemos hacer 2 observaciones.
-Si comparamos 2 atributos(ninguno es el atributo clase) es preferible que tengan una baja correlación 
-Si comparamos 1 atributo con el atributo clase es preferible que se tenga una alta correlación.

```python
#Coeficientes de correlación de Pearson
pd.set_option('display.width',100)
pd.set_option('precision',3)
correlation=data.corr(method='pearson')
print(correlation)
```

# <font color="#004D7F"> 4. Visualización univariable</font>

Las gráficas univariable nos permiten visualizar los atributos individuales sin interacciones; las cuales, el objetivo principal de las mismas es aprender algo sobre la distribución, la tendencia y la propagación de cada atributo. Aqui se hara presente la libreria matplotlib y seaborn. 

<a id="section21"></a>
## <font color="#004D7F">4.1. Histogramas </font>

Usamos esta gráfica conocida para poder conocer la distrbución de nuestros atributos, si es gausseano, sesgado a la izquierda o derecha, si tiene una distribución exponencial, etc. También es útil para ver valores atípicos. Primero definimos el tamaño de nuestros graficos con la función `figure(figsize=())` y luego hacemos uso de la funcion `hist()`


```python
%matplotlib inline
# Histogramas Univariados (Matplotlib)
import matplotlib.pyplot as ptl
fig = ptl.figure(figsize=(14,14))
ax=fig.gca()
data.hist(ax=ax)
ptl.show()
```

Vemos que no aparece el histograma para los atributos month(mes) y day(dia) puesto que son nombres ambos. Vemos que el atributo temp(temperatura) es el que se acerca más a una distribución gauseana, esto es intesante porque muchas técnicas de aprendizaje automático suponen una distribución univariada gaussiana en las variables de entrada. Vemos que que el atributo wind(viento) tiene cierto parecido con una gausseana y que el atributo RH se acerca a una gauseana pero con un sesgo hacia la derecha. 

Esto mismo lo podemos observar pero con la libreria Seaborn que es un paquete de visualización más profesional que Matplotlib, además de usar este por debajo. Podemos ver los Histogramas con una línea de densidad que nos muestra su distribución. Aqui igual que cuando graficamos solo con matplotlib primero definimos el tamaño de la matriz y las dimensiones de esta. Luego usamos la función `distplot` para cada atributo y en que parte de la matriz se encontrará (ambos estan como parametros)

```python
%matplotlib inline
# Histogramas univariados (Seaborn)
import matplotlib.pyplot as ptl
import seaborn as sns
f,axes=ptl.subplots(6,2,figsize=(20,20))
sns.distplot(data["DC"],ax=axes[0,0])
sns.distplot(data["DMC"],ax=axes[0,1])
sns.distplot(data["FFMC"],ax=axes[1,0])
sns.distplot(data["ISI"],ax=axes[1,1])
sns.distplot(data["RH"],ax=axes[2,0])
sns.distplot(data["X"],ax=axes[2,1])
sns.distplot(data["Y"],ax=axes[3,0])
sns.distplot(data["area"],ax=axes[3,1])
##sns.distplot(data["rain"],ax=axes[4,0])
sns.distplot(data["temp"],ax=axes[4,1])
sns.distplot(data["wind"],ax=axes[5,1])
```

Como se puede observar con Seaborn podemos identificar mejor la distribución de cada atributo.

<a id="section22"></a>
## <font color="#004D7F">4.2. Densidad </font>

Las gráficas se ven como un histograma abstracto con una curva suave dibujada a través de la parte superior de cada contenedor, al igual que su ojo intentó hacer con los histogramas. Podemos ver que la distribución de cada atributo es más clara que los histogramas. Aqui tambien primero definimos el tamaño y luego en vez de la función hist() usamos la funcion `plot()`, para el parametro 'kind' colocamos density y para 'layout' colocamos las dimensiones de la matriz en donde estaran los graficos   

```python
%matplotlib inline
# Densidad Univariada (Matplotlib)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(20,20))
ax=fig.gca()
data.plot(ax=ax, kind='density', subplots=True, layout=(6,2), sharex=False)
plt.show
```

Al igual que hemos visto antes podemos ver también el gráfico con la libreria Seaborn. De igua forma primero definimos el tamaño de la matriz y las dimensiones luego volveremos a usar la función `distplot` y al igual que cuando graficamos los histogramas con Seaborn ingresamos como parametros el atributo y la posición en la matriz donde se ubicará, pero hay algo adicional aqui, debemos agregarle el parametro hist (de histograma) y ponerle el valor False para que este no sea graficado y ponerle el parametro rug como True

```python
%matplotlib inline
# Densidad Univariada (Seaborn)
import matplotlib.pyplot as ptl
import seaborn as sns
f,axes=ptl.subplots(6,2,figsize=(20,20))
sns.distplot(data["DC"],hist=False,rug=True,ax=axes[0,0])
sns.distplot(data["DMC"],hist=False,rug=True,ax=axes[0,1])
sns.distplot(data["FFMC"],hist=False,rug=True,ax=axes[1,0])
sns.distplot(data["ISI"],hist=False,rug=True,ax=axes[1,1])
sns.distplot(data["RH"],hist=False,rug=True,ax=axes[2,0])
sns.distplot(data["X"],hist=False,rug=True,ax=axes[2,1])
sns.distplot(data["Y"],hist=False,rug=True,ax=axes[3,0])
sns.distplot(data["area"],hist=False,rug=True,ax=axes[3,1])
##sns.distplot(data["rain"],hist=False,rug=True,ax=axes[4,0])
sns.distplot(data["temp"],hist=False,rug=True,ax=axes[4,1])
sns.distplot(data["wind"],hist=False,rug=True,ax=axes[5,1])
```

Las rayas que aparecen en el eje X son los valores para cada atributo Para el atributo area(area incendiada) se observa que los valores estan muy pegados y la mayoria esta entre 0 y 50 aprox. Para el atributo DC también se observa valores pegados y la mayoria estan entre [0,100] y [580,780] aprox. . Para el atributo FFMC tenemos valores pegados también que se encuentran entre [85,100] aprox. Para el atributo RH vemos el caso contrario puesto que los valores se encuentran separados

También podemos hacer uso de los diagramas de cajas, aqui hacemos lo mismo que cuando graficamos la densidad con matplotlib, usamos la funcion plot con los mismos parametros, la diferencia esta que para el parametro 'kind' escribimos box en vez de density

## <font color="#004D7F">5.1. Matriz de correlación </font>

La correlación nos sirve para obervar que tanto se relacion los atributos entre si (incluido el atirbuto clase) como ya se explicó anteriormente. En esta ocasión dibujaremos una matriz con esos numeros con los siguientes parametros paso a paso como indica el código: tamaño de los cuadrantes, color de la matriz, valor minimo y maximo de los numeros que estaran en la matriz,de cuanto en cuanto van a saltar estos valores para la barra indicadora de la derecha, el eje x, el eje y y que nombres apareceran en ambos ejes. Vemos que debemos definir un arreglo con los nombres de los atributos. 
Los cuadrados serán más amarillos cuandos se acerquen al 1, más azules cuando se acerquen al -1 y neutral (entre celeste y verde) cuando se acerquen al 0.
También podemos ver que cada variable está perfectamente correlacionada positivamente entre sí (como era de esperar) en la línea diagonal desde la parte superior izquierda a la parte inferior derecha.

```python
import numpy as np
%matplotlib inline
# Matriz de correlación (Matplotlib)
correlations=data.corr()
fig=ptl.figure(figsize = (15,15))
ax=fig.add_subplot(111)
cax=ax.matshow(correlations,vmin=-1,vmax=1)
fig.colorbar(cax)
ticks=np.arange(0,11,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
names=['DC','DMC','FFMC','ISI','RH','X','Y','area','rain','temp','wind']
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

Al igual que hemos visto antes podemos ver también el gráfico con la librería Seaborn. Para ello usaremos la funcion `heatmap` en donde se deben colocar sus parametros

```python
%matplotlib inline
#Matriz de correlación (Seaborn)
correlation=data.corr()
plt.figure(figsize=(15,15))
ax=sns.heatmap(correlation,vmax=1,square=True,annot=True,cmap='viridis')
plt.title("Matriz de correlación")
plt.show
```

Vemos que los atributos que tienen más correlacion en sentido directo comparados con el resto son el DC con el DMC con un valor de 0.68 (aunque en sí sigue siendo bajo porque no es tan cercano al 1) esto quiere decir que si el DC aumenta o disminuye pasará lo mismo con el DMC7
Vemos que los atributos que tienen más relacion en sentido inverso comparados con el reston son el RH con el temp(temperatura) con un valor de  -0.53 (igual no es tan alta es correlación porque es algo lejana al -1), esto quiere decir que si el RH aumenta o disminuye la temp disminuira o aumentará.

## <font color="#004D7F">5.3. Matriz de dispersión </font>

Al igual que el diagrama de matriz de correlación anterior, la matriz de diagrama de dispersión es simétrica. Esto es útil para mirar las relaciones por pares desde diferentes perspectivas. Debido a que no tiene mucho sentido dibujar un diagrama de dispersión de cada variable consigo mismo, la diagonal muestra histogramas de cada atributo.
Si esque la dispersión entre 2 atributos tiene una forma cercana a una recta entonces seria un problema de regresión entre ellos. Si la dispersión se agrupan en algunas zonas del eje x e y entonces seria un problema de clusterización(agrupamiento) 

```python
%matplotlib inline
#Matriz de dispersión (Matplotlib)
from pandas.plotting import scatter_matrix
plt.rcParams["figure.figsize"]=[20,15]
scatter_matrix(data)
plt.show
```

Al igual que hemos visto antes podemos ver también el gráfico con la librería Seaborn. Aqui usaremos la función `pairplot`

```python
%matplotlib inline
# Matriz de dispersion (Seaborn)
sns.pairplot(data)
```

Entre los atributos DMC y DC hay cierta tendencia a una recta y se podria tratar de un problema de regresión. Para varios atributos al ir comparandose de 2 se observa cierto agrupamiento(clusterización). 

<a id="section33"></a>
## <font color="#004D7F">5.4.  Matriz de dispersion por clase </font>

Los puntos en una matriz de diagrama de dispersión pueden ser coloreados por la etiqueta de clase en problemas de clasificación. Esto puede ayudarnos a detectar una separación clara (o no clara) de las clases y quizás a darnos una idea de lo difícil que puede ser el problema.

Al igual que el anterior gráfico los diagramas están dados por la interacción de los pares de atributos pero, en este caso, teniendo en cuenta la etiqueta clase.

Usamos la funcion pairplot pero con más argumentos

```python
%matplotlib inline
# Matriz de dispersión por clase (Seaborn)
sns.pairplot(data,hue='area',diag_kind='hist')
```

Podemos observar que hay bastante espacio en el grafico, esto es porque al ser nuestro atributo clase numerico que se trabaja para un problema de regresión tiene distintos valores, pero observamos que la mayorias de los valores de nuestro atributi clase (area incendiada) estan entre 2.5 y 14.68
